````python
def calculate_variance(nl):
    """
    Calculate the population variance of a list of numbers.

    This function takes a list of numeric values as input and calculates the population variance
    using the formula: population variance = sum((x - mean)^2) / n

    Args:
        nl (list of float or int): A list of numeric values for which population variance will be calculated.

    Returns:
        float: The calculated population variance of the input list.

    Raises:
        TypeError: If the input list contains non-numeric values.
        ValueError: If the input list is empty.
    """
    if not all(isinstance(n, (int, float)) for n in nl):
        raise TypeError("Input list should contain only numeric values")

    if len(nl) == 0:
        raise ValueError("Input list cannot be empty")

    mean = math.fsum(nl) / len(nl)
    sum_squared_diff = math.fsum((n - mean) ** 2 for n in nl)
    population_variance = sum_squared_diff / len(nl)

    return population_variance
````
